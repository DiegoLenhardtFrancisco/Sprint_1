```python
import pandas as pd
import numpy as np
```

```python
#Importando dataset
df_ordem0 = pd.read_csv('Datasets/olist_order_items_dataset.csv', sep=',', encoding='iso-8859-1')
df_ordem1 = pd.read_csv('Datasets/olist_orders_dataset.csv', sep=',', encoding='iso-8859-1')
```

```python
#Associando os dois dataframe
df_ordem = pd.merge(df_ordem0, df_ordem1, on='order_id', how='inner')
df_ordem.head()
```

```python
#Peguei a coluna data limite de envio (shipping_limit_date) para trabalhar com datas
# Convertendo a coluna de datas para o tipo datetime, se necessário
df_ordem['order_approved_at'] = pd.to_datetime(df_ordem['order_approved_at'])

# Extraindo o ano da coluna de datas
df_ordem['ano'] = df_ordem['order_approved_at'].dt.year
```

```python
#Extraindo colunas nas quais vou trabalhar
df_ordem1 = df_ordem[['product_id', 'price', 'ano', 'seller_id']]
```

```python
#Eliminando dados repetidos
df1 = df_ordem1.drop_duplicates(subset='seller_id')
```

```python
#Criando dataframes por ano para compara-los uns com os outros
df_2016 = df1[df1['ano'] == 2016]
df_2017 = df1[df1['ano'] == 2017]
df_2018 = df1[df1['ano'] == 2018]
```

```python
#Relacionando os anos de 2017 e 2018 para averiguar se produtos foram vendidos em ambos os anos
df_2017_2018 = pd.merge(df_2017, df_2018, on='product_id', how='inner')
```

```python
#Percebemos que existem 10 produtos que foram vendidos nos respectivos anos
df_2017_2018.head(50)
```

```python
#Relacionando anos 2016 e 2017
df_2016_2017 = pd.merge(df_2016, df_2017, on='product_id', how='inner')
```

```python
#Percebemos que não teve produtos vendidos em ambos os anos
df_2016_2017
```

```python
#Relacionando anos de 2016 e 2018
df_2016_2018 = pd.merge(df_2016, df_2018, on='product_id', how='inner')
```

```python
# Preços em 2017 e 2018
precos_2017 = [35.00, 32.99, 259.60, 32.70, 26.00, 190.00, 214.00, 55.00, 85.00, 339.00]
precos_2018 = [34.50, 39.90, 278.99, 37.88, 22.50, 146.99, 214.00, 69.90, 104.90, 154.91]

# Calculando a variação de preço em R$
variacao_r = [precos_2018[i] - precos_2017[i] for i in range(len(precos_2017))]

# Calculando a variação de preço em porcentagem
variacao_percentual = [(variacao_r[i] / precos_2017[i]) * 100 for i in range(len(precos_2017))]

for i in range(len(precos_2017)):
    print(f'Produto {i+1}: Variação em R$ = {variacao_r[i]:.2f} | Variação percentual = {variacao_percentual[i]:.2f}%')
```
